### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
# 1. Create a Numpy array of shape (5,5) with random integers
array4 = np.random.randint(1, 51, size=(5, 5))
print("Array 4:")
print(array4)

# 1. Compute the mean, median, and standard deviation and variance of the array
mean = np.mean(array4)
median = np.median(array4)
std_dev = np.std(array4)
var = np.var(array4)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
print("Variance:", var)


# 2. Create a Numpy array of shape(3,3) with values from 1 to 9.
array5 = np.arange(1, 10).reshape(3, 3)
print("Array 5:")
print(array5)

# 2. Normalize the array (i.e scale the values to have a mean of 0 and a standard deviation of 1)
array5_normalized = (array5 - np.mean(array5)) / np.std(array5)
print("Normalized Array 5:")
print(array5_normalized)
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python
# 1. Create a Numpy array of shape (3,3) representing a matrix
array10 = np.random.randint(1, 51, size=(3, 3))
print("Array 10:")
print(array10)

# 1. Compute the determinant, inverse, eigenvalues, and transpose of the matrix
det = np.linalg.det(array10)
print("Determinant:", det)
try:
    inv = np.linalg.inv(array10)
    print("Inverse:")
    print(inv)
except np.linalg.LinAlgError:
    print("Inverse: Not computable (matrix may be singular)")
eigenvalues, eigenvectors = np.linalg.eig(array10)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
print("Transpose:")
print(array10.T)


# 2. Create two Numpy arrays of shape (2,3) and (3,2).
array11 = np.random.randint(1, 51, size=(2, 3))
print("Array 11:")
print(array11)
array12 = np.random.randint(1, 51, size=(3, 2))
print("Array 12:")
print(array12)


# 2. perform multiplication on these arrays 
result = np.dot(array11, array12)
print("Result of multiplication:")
print(result)
```

### Assignment 9: Structured Arrays

1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.
2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
# 1. Create a structured array with fields 'name'(String) and 'age' (integer), and 'height' (float).
dtype = np.dtype([('name', 'U10'), ('age', 'i4'), ('height', 'f4')])
array5 = np.array([('Alice', 25, 5.5), ('Bob', 30, 6.0), ('Charlie', 35, 5.8)], dtype=dtype)
print("Array 5:")
print(array5)

# 1. Add some data and sort the array by age.
new_data = np.array([('David', 28, 5.9)], dtype=dtype)
array5 = np.append(array5, new_data)
array5 = np.sort(array5, order='age')
print("Names in Array 5:")
print(array5['name'])


# 2. Create a structured array with fields 'X' and 'Y' (both integers).
dtype2 = np.dtype([('X', 'i4'), ('Y', 'i4')])
array6 = np.array([(1, 2), (3, 4), (5, 6)], dtype=dtype2)
print("Array 6:")
print(array6)

# 2. Add some data and compute the Euclidean distance between each pair of points.
new_points = np.array([(7, 8), (9, 10)], dtype=dtype2)
array6 = np.append(array6, new_points)
print("Array 6 after adding new points:")
print(array6)

# Compute Euclidean distance between each pair of points
def compute_distance(point1, point2):
    return np.sqrt((point1['X'] - point2['X'])**2 + (point1['Y'] - point2['Y'])**2)

distances = []
for i in range(len(array6)):
    for j in range(i + 1, len(array6)):
        dist = compute_distance(array6[i], array6[j])
        distances.append((array6[i], array6[j], dist))

print("Euclidean distances between points:")
for p1, p2, dist in distances:
    print(f"Distance between {p1} and {p2} is {dist}")
    
```
